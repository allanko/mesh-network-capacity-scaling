```python
%matplotlib inline
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 13 , 10

import glob, os
import pandas as pd
import numpy as np
import datetime
```

```python
# import data 
# count up connectivity matrix by counting number of packets exchanged by each node pair
```

```python
collectdir = r'170424/deployed/n40/n40_all_calibration_10_1704241420/' # summary.xlsx file of the experiment
n = 40  # number of nodes in experiment
minutes_in_exp = 4 
attempted_packets_per_sec = 10

expected_total_packets_attempted = 60*minutes_in_exp*attempted_packets_per_sec

mgenlogs = pd.read_excel(os.path.join(collectdir, 'summary.xlsx'), sheetname = 'MGENlogs')
mgenlogs['src'] = [int(i[7:-5]) for i in mgenlogs['src']]
mgenlogs['dst'] = [int(i[7:-5]) for i in mgenlogs['dst']]
```

```python
zerofill = np.zeros(shape=(n,n))
connectivity = pd.DataFrame(zerofill,
                               columns = [i+1 for i in range(n)],
                               index = [i+1 for i in range(n)])

# a square matrix, indexed by column, index
# connectivity[x][y] accesses column x, row y
# let x be source node, y be destination node
# add 1 to cells via connectivity.set_value(y, x, connectivity[x][y] + 1)
```

```python
for i in mgenlogs[['dst','src']].index.values:
    x = mgenlogs['src'][i]
    y = mgenlogs['dst'][i]
    connectivity.set_value(y, x, (connectivity[x][y] + 1))

connectivity_norm = connectivity/expected_total_packets_attempted
```

```python
connectivity_norm
```

```python
# plot as a heatmap
```

```python
fig = plt.imshow(connectivity, interpolation='nearest', cmap='plasma', vmin = 0)
ax = fig.axes

_ = ax.set_xticks(np.linspace(0, n-1, n))
_ = ax.set_yticks(np.linspace(0, n-1, n))
_ = ax.set_xticklabels(connectivity.columns, fontsize = 10)
_ = ax.set_yticklabels(connectivity.index, fontsize = 10)
_ = ax.set_xlabel('source node', fontsize=20)
_ = ax.set_ylabel('destination node', fontsize=20)

ax.grid('off')
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top') 

cb = plt.colorbar()
cb.set_label('packets received', fontsize=15)
cb.ax.tick_params(labelsize=15)

# write to file - 
plt.savefig(os.path.join(collectdir, 'connectivity-counts.png'), bbox_inches='tight')
```

```python
fig = plt.imshow(connectivity_norm, interpolation='nearest', cmap='plasma', vmin = 0, vmax = 1)
ax = fig.axes

_ = ax.set_xticks(np.linspace(0, n-1, n))
_ = ax.set_yticks(np.linspace(0, n-1, n))
_ = ax.set_xticklabels(connectivity.columns, fontsize = 10)
_ = ax.set_yticklabels(connectivity.index, fontsize = 10)
_ = ax.set_xlabel('source node', fontsize=20)
_ = ax.set_ylabel('destination node', fontsize=20)

ax.grid('off')
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top') 

cb = plt.colorbar()
cb.set_label('fraction of packets received', fontsize=15)
cb.ax.tick_params(labelsize=15)

# write to file - 
plt.savefig(os.path.join(collectdir, 'connectivity.png'), bbox_inches='tight')
```
